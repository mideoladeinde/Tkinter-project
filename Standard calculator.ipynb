```python
import tkinter as tk

def calcu():
    top = tk.Tk()

    dsp_entry = tk.Entry(top)
    dsp_entry.grid(row = 0, column = 0, columnspan = 4)

    def button_click(number):
        val = dsp_entry.get()
        dsp_entry.delete(0, tk.END)
        dsp_entry.insert(0, str(val) + str(number))

    def button_clear():
        dsp_entry.delete(0, tk.END)
        
    def button_equals():
        try:
            result = eval(dsp_entry.get())
            dsp_entry.delete(0, tk.END)
            dsp_entry.insert(0, str(result))
        except Exception as e:
            dsp_entry.delete(0, tk.END)
            dsp_entry.insert(0, "Error")


    button_no7 = tk.Button(top, text = '7', command = lambda: button_click(7))
    button_no7.grid(row = 1, column = 0)

    button_no8 = tk.Button(top, text = '8', command = lambda: button_click(8))
    button_no8.grid(row = 1, column = 1)

    button_no9 = tk.Button(top, text = '9', command = lambda: button_click(9))
    button_no9.grid(row = 1, column = 2)

    button_div = tk.Button(top, text = 'รท', command = lambda: button_click('/'))
    button_div.grid(row = 1, column = 3)

    button_no4 = tk.Button(top, text = '4', command = lambda: button_click(4))
    button_no4.grid(row = 2, column = 0)

    button_no5 = tk.Button(top, text = '5', command = lambda: button_click(5))
    button_no5.grid(row = 2, column = 1)

    button_no6 = tk.Button(top, text = '6', command = lambda: button_click(6))
    button_no6.grid(row = 2, column = 2)

    button_multi = tk.Button(top, text = 'ร', command = lambda: button_click('*'))
    button_multi.grid(row = 2, column = 3)

    button_no1 = tk.Button(top, text = '3', command = lambda: button_click(3)) 
    button_no1.grid(row = 3, column = 0)

    button_no2 = tk.Button(top, text = '2', command = lambda: button_click(2))
    button_no2.grid(row = 3, column = 1)

    button_no3 = tk.Button(top, text = '1', command = lambda: button_click(1))
    button_no3.grid(row = 3, column = 2)

    button_sub = tk.Button(top, text = '-', command = lambda: button_click('-'))
    button_sub.grid(row = 3, column = 3)

    button_dot = tk.Button(top, text = '.', command = lambda: button_click('.'))
    button_dot.grid(row = 4, column = 0)

    button_zero = tk.Button(top, text = '0', command = lambda: button_click(0))
    button_zero.grid(row = 4, column = 1)

    call_result = tk.Button(top, text = '=', command = button_equals)
    call_result.grid(row = 4, column = 2)

    button_add = tk.Button(top, text = '+', command = lambda: button_click('+'))
    button_add.grid(row = 4, column = 3)

    button_clear = tk.Button(top, text='C', command=button_clear)
    button_clear.grid(row=5, column=0, columnspan=4)

    top.mainloop()

calcu()
```
