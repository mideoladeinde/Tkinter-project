```python
import datetime
import tkinter as tk
from tkinter import messagebox

def ScheduleTask():

    top = tk.Tk()

    header = tk.Label(top, text = 'SCHEDULE TASK!', bg = 'grey', font = ('Calibre', 15, 'bold'))
    header.grid(row = 0, column = 0, columnspan = 4, padx = 100, pady = 20)

    task_label = tk.Label(top, text = 'Task:', font = ('Calibre', 10, 'bold'))
    task_label.grid(row = 1, column = 0)
    task_entry = tk.Entry(top)
    task_entry.grid(row = 2, column = 0)

    year_label = tk.Label(top, text = 'Year:', font = ('Calibre', 10, 'bold'))
    year_label.grid(row = 3, column = 0)
    year_entry = tk.Entry(top)
    year_entry.grid(row = 4, column = 0)

    month_label = tk.Label(top, text = 'Month:', font = ('Calibre', 10, 'bold'))
    month_label.grid(row = 5, column = 0)
    month_entry = tk.Entry(top)
    month_entry.grid(row = 6, column = 0)

    day_label = tk.Label(top, text = 'Day:', font = ('Calibre', 10, 'bold'))
    day_label.grid(row = 7, column = 0)
    day_entry = tk.Entry(top)
    day_entry.grid(row = 8, column = 0)

    hour_label = tk.Label(top, text = 'Hour:', font = ('Calibre', 10, 'bold'))
    hour_label.grid(row = 9, column = 0)
    hour_entry = tk.Entry(top)
    hour_entry.grid(row = 10, column = 0)

    create_button = tk.Button(top, text = 'Create Schedule.', bg = 'grey', command = lambda: create_button(task_entry, year_entry, month_entry, day_entry, hour_entry, txt_file))
    create_button.grid(row = 11, column = 0, pady = 10)

    viewTask_button = tk.Button(top, text = 'View Tasks.', bg = 'grey', command = lambda: viewTask_button(txt_file, viewTask_label))
    viewTask_button.grid(row = 1, column = 5, padx = 30)

    view_frame = tk.Frame(top, bd = 80, bg = 'white')
    view_frame.grid(row = 2, column = 5, rowspan = 10, columnspan = 10, padx = 10, pady = 10)
    viewTask_label = tk.Label(view_frame, text = 'Pending.')
    viewTask_label.grid(row = 0, column = 0, padx = 10, pady = 10)


    txt_file = open('Scheduled_task.txt','wt')


    def create_button(task_entry, year_entry, month_entry, day_entry, hour_entry, txt_file):

        schedule_data = None

        user_data = {
            'Task': task_entry.get(),
            'Year': year_entry.get(),
            'Month': month_entry.get(),
            'Day': day_entry.get(),
            'Hour': hour_entry.get()
        }
        while True:
            try:
                schedule_data = datetime.datetime(int(user_data.get('Year')), int(user_data.get('Month')), int(user_data.get('Day')), int(user_data.get('Hour')))
            except:
                messagebox.showinfo('Error', 'Make sure the fields ae filled in and the correct data is privided.')
                break

            if(schedule_data < datetime.datetime.now()):
                messagebox.showinfo('Error', 'Please input a valid date for your schedule.')
                break

            txt_file.write(f'Task :\t\t{user_data.get('Task')}\nYear :\t\t{schedule_data.year}\nMonth :\t\t{schedule_data.month}\nDay :\t\t{schedule_data.day}\nHour :\t\t{schedule_data.hour}\n\n')

            messagebox.showinfo('Success', 'Your task schedule has been successfully saved!')
            break

    def viewTask_button(txt_file, viewTask_label):
        try:
            txt_file.close()
            txt_file = open('Scheduled_task.txt', 'r')
            file_prop = txt_file.read()
            txt_file.close()
            txt_file = open('Scheduled_task.txt', 'at')
            
            if len(file_prop) < 1:
                viewTask_label.config(text = f'No Schedule has been saved yet')

            else:
                viewTask_label.config(text = f'Task list\n{file_prop}')
                
        except Exception as e:
            messagebox.showinfo('Error', str(e))

    def close(txt_file):
        txt_file.close()

    top.mainloop()

    close(txt_file)

ScheduleTask()
```
